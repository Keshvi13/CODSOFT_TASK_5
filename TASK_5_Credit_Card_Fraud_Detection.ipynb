# DATA ANALYSIS

In this part of the project I will visualize some of my data to check for relationships and trends in the data

```python
# I'll import seaborn visualization library 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def draw_histogram(column,df=df.head(1000)):
    """I will draw a sample of 1000 and make some statistical inferences,this function
    returns a plotted histogram.The function is based on the seaborn histogram documentation"""
    fig = plt.figure(figsize=(20,8))
    sns.histplot(x=column, data=df)
    plt.xlabel(column)
    plt.ylabel("Total in Frequency")
    plt.title(f"The Histogram Distribution of {column}")
    plt.show()
    
```

```python
for column in df.columns:
    if column in ['Time','Amount']:
        draw_histogram(column)
    else :
        pass
```

```python
# I'll draw a countplot for my Targets ,to have an idea of what my outcomes might be
fig = plt.figure(figsize=(20,8))
sns.countplot(x='Class',data=df)
plt.xlabel('Fraud Class')
plt.ylabel("Total Number")
plt.title("How Transactions have been classified")
plt.show()
```
